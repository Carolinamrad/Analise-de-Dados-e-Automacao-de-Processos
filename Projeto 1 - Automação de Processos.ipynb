```python
# Passo 4: Importar os dados e Calcular os indicadores
import pandas as pd # apelido para evitar repetir o nome, muito utilizado no mercado

# importar a base de dados para uma tabela
tabela = pd.read_excel(r"C:\Users\marcos.pardo\Downloads\Vendas - Dez.xlsx")
display(tabela)

# cálculo dos indicadores
total_gasto = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto/quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)

# caso o arquivo esteja em formato csv com separadores deve se usar a sintaxe abaixo para carregar e separar as colunas
# tabela = pandas.read_csv(r"C:\Users\marcos.pardo\Downloads\Vendas - Dez.xlsx", sep=";")
```
